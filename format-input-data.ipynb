```python
import os
import pandas as pd
```

```python
filenames = os.listdir('input')
```

```python
with open('raw.csv', 'w') as outfile:
    for fname in filenames:
        with open(os.path.join('input', fname)) as infile:
            for line in infile:
                outfile.write(line)
```

```python
data = []
with open('raw.csv') as f:
    for line in f:
        data.append(line.split('|'))
```

```python
data
```

```python
questions = list(set([x[0] for x in data]))
```

```python
answers = []
for entry in data:
    answers.append((questions.index(entry[0]), entry[1].replace('\n', ''), None))
```

```python
questions = [q.replace('\n', '') for q in questions]
```

```python
list(enumerate(questions))
```

```python
ans_headers = ['Q_ID', 'ANSWER', 'CORRECT']
que_headers = ['QUESTIOIN']
```

```python
ans_pd = pd.DataFrame(data=answers, columns=ans_headers)
que_pd = pd.DataFrame(data=questions, columns=que_headers)
```
